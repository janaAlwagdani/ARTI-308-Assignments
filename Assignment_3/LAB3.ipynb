# 1. Import Libraries

```python
# Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots look cleaner
sns.set()
```

# 2. Loading data

```python
# Load Dataset

df = pd.read_csv("Daily_Water_Intake.csv")

# Display first 5 rows
df.head(7)
```

# 3. Data type of columns

```python
# viewing the data types of columns
df.dtypes
```

```python
df.info()
```

# 4. Check Missing Values

```python
# Check Missing Values
print(df.isna().sum())
```

# 5. Check duplicate rows

```python
# Check duplicate rows
df.duplicated()[df.duplicated()==True]
```

# 6. No. of rows and columns

```python
# finding number of rows and columns

print("Shape (rows, columns): ", df.shape,"\n")
print("number of rows: ", df.shape[0])
print("number of columns: ", df.shape[1])
```

# 7. Descriptive summary Statistics

```python
# Statistical summary
df.describe(include='all')
```

```python
cat_cols = ["Gender", "Physical Activity Level", "Weather", "Hydration Level"]

for col in cat_cols:
    print("\n", col)
    print(df[col].value_counts())
```

# 8. Univariate Analysis

```python
#Age
plt.figure()
df["Age"].hist()
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
#Weight
plt.figure()
df["Weight (kg)"].hist()
plt.title("Weight Distribution")
plt.xlabel("Weight (kg)")
plt.ylabel("Count")
plt.show()
```

```python
#Daily Water Intake
plt.figure()
df["Daily Water Intake (liters)"].hist()
plt.title("Water Intake Distribution")
plt.xlabel("Liters")
plt.ylabel("Count")
plt.show()
```

```python
#Gender
df["Gender"].value_counts().plot(kind="bar")
plt.title("Gender Distribution")
plt.show()
```

```python
#Physical Activity
df["Physical Activity Level"].value_counts().plot(kind="bar")
plt.title("Activity Level Distribution")
plt.show()
```

```python
#Weather
df["Weather"].value_counts().plot(kind="bar")
plt.title("Weather Distribution")
plt.show()
```

```python
#Hydration Level
df["Hydration Level"].value_counts().plot(kind="bar")
plt.title("Hydration Level Distribution")
plt.show()
```

---
Univariate analysis was conducted to understand the distribution and characteristics of each feature individually using histograms for numerical variables and bar charts for categorical variables.

----

# 9. Bivariate Analysis


```python
# Age vs Water Intake

plt.figure()
plt.scatter(df["Age"], df["Daily Water Intake (liters)"])
plt.xlabel("Age")
plt.ylabel("Water Intake (liters)")
plt.title("Age vs Water Intake")
plt.show()
```

```python
# Weight vs Water Intake

plt.figure()
plt.scatter(df["Weight (kg)"], df["Daily Water Intake (liters)"])
plt.xlabel("Weight (kg)")
plt.ylabel("Water Intake (liters)")
plt.title("Weight vs Water Intake")
plt.show()
```

```python
# Gender vs Water Intake

plt.figure()
df.boxplot(column="Daily Water Intake (liters)", by="Gender")
plt.title("Water Intake by Gender")
plt.suptitle("")
plt.show()
```

```python
# Activity Level vs Water Intake

plt.figure()
df.boxplot(column="Daily Water Intake (liters)", by="Physical Activity Level")
plt.title("Water Intake by Activity Level")
plt.suptitle("")
plt.show()
```

```python
# Weather vs Water Intake

plt.figure()
df.boxplot(column="Daily Water Intake (liters)", by="Weather")
plt.title("Water Intake by Weather")
plt.suptitle("")
plt.show()
```

```python
# Correlation Matrix
plt.figure()
sns.heatmap(df[["Age","Weight (kg)","Daily Water Intake (liters)"]].corr(),
            annot=True)
plt.title("Correlation Matrix")
plt.show()
```

---
Bivariate analysis was performed to examine relationships between variables using scatter plots for numerical features and boxplots for categorical versus numerical comparisons.

# 10. Time-Based Analysis



Time-based analysis was not conducted in this study because the dataset does not include any temporal features such as date, time, or timestamps. The available variables represent static demographic, environmental, and behavioral factors measured at a single point rather than over a period. Therefore, the analysis focused on cross-sectional relationships between variables instead of temporal trends.

-----

# Summary

In this exploratory data analysis, the dataset was examined to understand the distribution, relationships, and key patterns among the variables related to daily water intake. The analysis showed that weight, physical activity level, and weather conditions are the most influential factors affecting water consumption. Higher physical activity and hotter weather were associated with increased water intake, while gender differences were minimal. Correlation analysis also indicated a strong positive relationship between weight and daily water intake. Overall, the dataset was clean with no missing values, making it suitable for further machine learning modeling.
